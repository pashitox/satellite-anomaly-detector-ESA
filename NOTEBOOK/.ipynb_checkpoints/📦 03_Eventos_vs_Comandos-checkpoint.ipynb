```python
import os
import pandas as pd

# Ruta a la carpeta de telecommands
tele_path = base_path + 'telecommands/'

# Lista de archivos CSV en la carpeta
csv_files = [f for f in os.listdir(tele_path) if f.endswith('.csv')]

# Resultado para guardar archivos con columnas de tipo fecha
detected_datetime_cols = {}

for fname in csv_files:
    try:
        df = pd.read_csv(os.path.join(tele_path, fname), nrows=5)  # carga solo las primeras filas
        for col in df.columns:
            sample = pd.to_datetime(df[col], errors='coerce')
            if sample.notna().sum() >= 1:  # Si al menos una fila es reconocida como datetime
                detected_datetime_cols[fname] = col
                break  # sólo guardamos la primera columna que parece de tiempo
    except Exception as e:
        print(f"⚠️ Error leyendo {fname}: {e}")

# Mostramos resultados encontrados
print("📅 Archivos con columnas de fecha detectadas:")
for f, c in detected_datetime_cols.items():
    print(f"🗂️ {f} → columna: {c}")
```
